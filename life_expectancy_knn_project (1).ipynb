Install all necessary packages.

```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

Import and clean data.

```python
df = pd.read_csv("countries_clean.csv")
```

```python
# remove all columns except year, SP.DYN.LE00.IN, income
df = df[["year", "SP.DYN.LE00.IN", "income"]]
```

```python
# rename SP.DYN.LE00.IN to life_expectancy
df = df.rename(columns = {"SP.DYN.LE00.IN": "life_expectancy"})
```

```python
df.head()
```

Label X and Y

```python
X = df.iloc[:, 0:2]
```

```python
y = df.iloc[:,2] 
```

Assign training and test data, 30% test data.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 11, test_size = 0.3)
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors = 15)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

Testing model

```python
knn.score(X_test, y_test)
```

Unfortunately a relatively low accuracy of 64%. Changing the number of nearest neighbours and the test data size did not improve upon this score.
